1.1. Create a function for template matching applying euclidean distance and normalized cross-correlation and display the results in this format:
    
<img src="images_notebook/tm.png" width="400" height="120">

**Use titles in all figures to understand what is being displayed.**

```python
# Your solution here
def visualize_image_3(tigre_1,tigre_2,tigre_3,label1,label2,label3):
    #Visualize 3 images
    fig=plt.figure(figsize=(12,12))
    fig.add_subplot(1,3,1)
    plt.title(label1)
    plt.imshow(tigre_1, cmap='gray')
    plt.xticks([])
    plt.yticks([])
    fig.add_subplot(1,3,2)
    plt.title(label2)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(tigre_2, cmap='gray')
    fig.add_subplot(1,3,3)
    plt.title(label3)
    plt.imshow(tigre_3, cmap='gray')
    plt.xticks([])
    plt.yticks([])


def template_matching(image, template, threshold_e, threshold_c):
    
    if len(template.shape)==3:
        if template.shape[2]==4: #Converts from RGBA to RGB if necessary
             template = rgba2rgb(template)
         
    if len(image.shape)==3:
        if image.shape[2]==4:
             image = rgba2rgb(image)
    
    image = rgb2gray(image) #Convert images to gray scale and then to float.
    template = rgb2gray(template)  
    image_f = img_as_float(image)
    template_f = img_as_float(template)
    
    #Euclidean distance
    dimensions = [image_f.shape[i] - template_f.shape[i] for i in range(2)] #Get difference of sizes from both images.
    euclid = np.ones(dimensions) #Create matrix from differences.
        
    for i in range(dimensions[0]): #For all pixels of image.
        for j in range(dimensions[1]):            
            region = image[i: i + template_f.shape[0], j: j + template_f.shape[1]] - template_f#Apply filter to region of image.
            euclid[i][j] = np.linalg.norm(region)
                
    euclid_t = euclid < threshold_e  #Lower values than threshold indicate where the template is.
    
    #Cross-correlation
    cross_correlation = match_template(image_f, template_f) #Apply cross-correlation filter.
    cross_correlation_t = cross_correlation > threshold_c #Higher values than threshold indicate where the template is.
    
    #Visualize results
    visualize_image_3(image, euclid, cross_correlation,"original image","euclidean distance","normalized cross-correlation")
    visualize_image_3(template, euclid_t, cross_correlation_t,"template","thresholded euclidean distance","thresholded normalized cross-correlation")
    

template_matching(einstein, eye, 7, 0.65)
```
